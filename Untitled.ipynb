# Dimensionality Reduction

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)  # Taking more components initially
pca_result = pca.fit_transform(df.iloc[:, :-1])
```

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
print("Explained Variance by Each Principal Component:")
for i, var in enumerate(explained_variance, 1):
    print(f"PC{i}: {var:.4f} ({cumulative_variance[i-1]:.4f} cumulative)")
```

```python
pca = PCA(n_components=5)
pca_result = pca.fit_transform(df.iloc[:, :-1])
explained_var = pca.explained_variance_ratio_
print("\nExplained Variance by PCA Components:", explained_var)
```

```python
# PCA Visualization
plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=df.iloc[:, -1], cmap='viridis', alpha=0.7)
plt.colorbar(label="Target Value")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Visualization of Spectral Features")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, pca.n_components + 1), pca.explained_variance_ratio_, marker='o', linestyle='--')
plt.xlabel("Principal Component")
plt.ylabel("Variance Explained")
plt.title("Explained Variance by Each Principal Component")
plt.show()
```

```python
# Cumulative Variance Plot
plt.figure(figsize=(8, 5))
plt.plot(range(1, pca.n_components + 1), np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Cumulative Variance Explained by PCA")
plt.show()
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
tsne_result = tsne.fit_transform(df.iloc[:, :-1])
plt.figure(figsize=(8, 6))
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=df.iloc[:, -1], cmap='coolwarm', alpha=0.7)
plt.colorbar(label="Target Value")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.title("t-SNE Visualization of Spectral Features")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_result[:, 0], pca_result[:, 1], pca_result[:, 2], c=df.iloc[:, -1], cmap='plasma', alpha=0.7)
ax.set_xlabel("Principal Component 1")
ax.set_ylabel("Principal Component 2")
ax.set_zlabel("Principal Component 3")
ax.set_title("3D PCA Visualization of Spectral Features")
plt.show()
```

# Splitting the dataset into Training and Testing set

```python
X = pca_result  # Features from PCA
y = df["vomitoxin_ppb"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(pca_result, df.iloc[:, -1], test_size=0.2, random_state=42)
```

# Training RandomForestRegressor

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

# Evaluation metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R² Score: {r2:.4f}")
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.xlabel("Actual DON Concentration")
plt.ylabel("Predicted DON Concentration")
plt.title("Actual vs Predicted DON Concentration")
plt.show()
```

# Hyperparameter tuning for Random Forest

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=5, n_jobs=-1, scoring='neg_mean_absolute_error')
grid_search.fit(X_train, y_train)
```

```python
best_rf = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)
```

```python
best_rf.fit(X_train, y_train)
y_pred = best_rf.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R² Score: {r2:.4f}")
```

# Training XGBosst

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(n_estimators=300, learning_rate=0.05, max_depth=6, random_state=42)
xgb.fit(X_train, y_train)
```

```python
y_pred_xgb = xgb.predict(X_test)
```

```python
mae_xgb = mean_absolute_error(y_test, y_pred_xgb)
rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
r2_xgb = r2_score(y_test, y_pred_xgb)

print("\nXGBoost Performance:")
print(f"Mean Absolute Error (MAE): {mae_xgb:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_xgb:.4f}")
print(f"R² Score: {r2_xgb:.4f}")
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(y_test, y_pred_xgb, alpha=0.6, color='red')
plt.xlabel("Actual DON Concentration")
plt.ylabel("Predicted DON Concentration (XGBoost)")
plt.title("Actual vs Predicted DON Concentration - XGBoost")
plt.show()
```

# Hyperparameter tuning for XGBoost

```python
param_grid_xgb = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 6, 10],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.7, 0.9, 1.0]
}

grid_search_xgb = GridSearchCV(XGBRegressor(random_state=42), param_grid_xgb, cv=5, n_jobs=-1, scoring='neg_mean_absolute_error')
grid_search_xgb.fit(X_train, y_train)
```

```python
best_xgb = grid_search_xgb.best_estimator_
print("Best Parameters for XGBoost:", grid_search_xgb.best_params_)
best_xgb.fit(X_train, y_train)
y_pred_xgb = best_xgb.predict(X_test)
mae_xgb = mean_absolute_error(y_test, y_pred_xgb)
rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
r2_xgb = r2_score(y_test, y_pred_xgb)

print("\nTuned XGBoost Performance:")
print(f"Mean Absolute Error (MAE): {mae_xgb:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_xgb:.4f}")
print(f"R² Score: {r2_xgb:.4f}")
```

# Training CNN model

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Conv1D, Flatten
from tensorflow.keras.optimizers import Adam
```

# Reshape data for CNN

```python
X_train_cnn = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test_cnn = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

# Building CNN Model

```python
cnn_model = Sequential([
    Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], 1)),
    Flatten(),
    Dense(32, activation='relu'),
    Dense(1)
])

cnn_model.compile(optimizer=Adam(learning_rate=0.01), loss='mse', metrics=['mae'])
```

# Training CNN model

```python
cnn_history = cnn_model.fit(X_train_cnn, y_train, validation_data=(X_test_cnn, y_test), epochs=100, batch_size=16, verbose=1)
```

```python
y_pred_cnn = cnn_model.predict(X_test_cnn).flatten()
```

```python
mae_cnn = mean_absolute_error(y_test, y_pred_cnn)
rmse_cnn = np.sqrt(mean_squared_error(y_test, y_pred_cnn))
r2_cnn = r2_score(y_test, y_pred_cnn)

print("\nCNN Performance:")
print(f"Mean Absolute Error (MAE): {mae_cnn:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_cnn:.4f}")
print(f"R² Score: {r2_cnn:.4f}")
```

# Reshape data for CNN

```python
X_train_cnn = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test_cnn = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

```python
# Build Improved CNN Model
from tensorflow.keras.layers import BatchNormalization, MaxPooling1D
cnn_model = Sequential([
    Conv1D(filters=128, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], 1), padding='same'),
    BatchNormalization(),
    MaxPooling1D(pool_size=2),
    Conv1D(filters=64, kernel_size=3, activation='relu', padding='same'),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(1)
])

cnn_model.compile(optimizer=Adam(learning_rate=0.005), loss='mse', metrics=['mae'])
```

```python
cnn_history = cnn_model.fit(X_train_cnn, y_train, validation_data=(X_test_cnn, y_test), epochs=150, batch_size=32, verbose=1)
```

```python
# Predictions using Improved CNN
y_pred_cnn = cnn_model.predict(X_test_cnn).flatten()
```

# Evaluate Improved CNN Model

```python
mae_cnn = mean_absolute_error(y_test, y_pred_cnn)
rmse_cnn = np.sqrt(mean_squared_error(y_test, y_pred_cnn))
r2_cnn = r2_score(y_test, y_pred_cnn)

print("\nImproved CNN Performance:")
print(f"Mean Absolute Error (MAE): {mae_cnn:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_cnn:.4f}")
print(f"R² Score: {r2_cnn:.4f}")
```

# Comparision of RandomForest, XGBoost and CNN Model

```python
print("\nComparison of Random Forest, XGBoost, MLP, and Improved CNN:")
print("{:<20} {:<15} {:<15} {:<15}".format("Model", "MAE", "RMSE", "R² Score"))
print("{:<20} {:<15.4f} {:<15.4f} {:<15.4f}".format("Random Forest", mae, rmse, r2))
print("{:<20} {:<15.4f} {:<15.4f} {:<15.4f}".format("Tuned XGBoost", mae_xgb, rmse_xgb, r2_xgb))
print("{:<20} {:<15.4f} {:<15.4f} {:<15.4f}".format("Improved CNN", mae_cnn, rmse_cnn, r2_cnn))
```
